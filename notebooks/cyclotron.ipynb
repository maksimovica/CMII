```python
import numpy as np

def get_effective_mass(theta, mt, ml):
    inverse = (np.cos(theta))**2 / mt**2 + (np.sin(theta))**2 / (ml*mt)
    return np.sqrt(1.0/inverse)
```

```python
import matplotlib.pyplot as plt
theta = np.array([i*np.pi/100.0/2.0 for i in range(100)])
fig, ax = plt.subplots(figsize = (10, 10), dpi=100, frameon=False)
ax.plot(theta*180.0/np.pi, get_effective_mass(theta, 0.97, 0.19))
```
